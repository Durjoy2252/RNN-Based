```python
import zipfile

zip_file_path = '/content/ESSg_dataset (3).zip'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall('.')

print(f"'{zip_file_path}' extracted successfully.")
```

```python
import os
import numpy as np

dataset_path = './ESSg_dataset/'
all_data = []
file_names = sorted(os.listdir(dataset_path), key=lambda x: int(os.path.splitext(x)[0])) # Sort by number to maintain order

for filename in file_names:
    if filename.endswith('.txt'):
        filepath = os.path.join(dataset_path, filename)
        with open(filepath, 'r') as f:

            content = f.read().strip()
            if content:
                try:

                    data_points = [float(x) for x in content.split()]
                    all_data.append(np.array(data_points))
                except ValueError as e:
                    print(f"Warning: Could not parse data in {filename}: {e}")
            else:
                print(f"Warning: {filename} is empty.")

print(f"Loaded {len(all_data)} files from '{dataset_path}'")

if all_data:
    for i in range(min(3, len(all_data))):
        print(f"Shape of data from {file_names[i]}: {all_data[i].shape}")
        print(f"First 5 elements of data from {file_names[i]}: {all_data[i][:5]}")

```

```python
import os

dataset_path = './ESSg_dataset/'
all_text_data = []
file_names = sorted(os.listdir(dataset_path), key=lambda x: int(os.path.splitext(x)[0]))

for filename in file_names:
    if filename.endswith('.txt'):
        filepath = os.path.join(dataset_path, filename)
        try:
            with open(filepath, 'r', encoding='utf-8-sig') as f:
                content = f.read().strip()
                if content:
                    all_text_data.append(content)
                else:
                    print(f"Warning: {filename} is empty.")
        except Exception as e:
            print(f"Error reading {filename}: {e}")

print(f"Loaded {len(all_text_data)} text files from '{dataset_path}'")

if all_text_data:
    for i in range(min(3, len(all_text_data))):
        print(f"\n--- Content from file {file_names[i]} ---")
        print(all_text_data[i][:200])
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


num_words = 10000
oov_token = "<unk>"
tokenizer = Tokenizer(num_words=num_words, oov_token=oov_token)


tokenizer.fit_on_texts(all_text_data)

print(f"Vocabulary size: {len(tokenizer.word_index)}")
print(f"Number of words considered (num_words): {num_words}")

sequences = tokenizer.texts_to_sequences(all_text_data)


sequence_lengths = [len(seq) for seq in sequences]


max_seq_len = max(sequence_lengths)
mean_seq_len = np.mean(sequence_lengths)
median_seq_len = np.median(sequence_lengths)

print(f"\nMax sequence length: {max_seq_len}")
print(f"Mean sequence length: {mean_seq_len:.2f}")
print(f"Median sequence length: {median_seq_len}")


maxlen = min(max_seq_len, 256)

print(f"Using maxlen for padding: {maxlen}")


padded_sequences = pad_sequences(sequences, maxlen=maxlen, padding='post', truncating='post')



print(f"\nShape of padded_sequences: {padded_sequences.shape}")
print(f"Example of a padded sequence (first 10 elements): {padded_sequences[0][:10]}")
print(f"Example of a padded sequence (last 10 elements): {padded_sequences[0][-10:]}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense


vocab_size = len(tokenizer.word_index) + 1


input_length = maxlen

embedding_dim = 128


model = Sequential()


model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=input_length))


model.add(LSTM(units=64))


model.add(Dense(units=1, activation='sigmoid'))


model.summary()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense


vocab_size = len(tokenizer.word_index) + 1


embedding_dim = 128


model = Sequential()


model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim))


model.add(LSTM(units=64))


model.add(Dense(units=1, activation='sigmoid'))


model.build(input_shape=(None, maxlen))


model.summary()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split


num_samples = padded_sequences.shape[0]
y_labels = np.zeros((num_samples, 1))
y_labels[num_samples // 2:] = 1

print(f"Shape of y_labels: {y_labels.shape}")
print(f"First 5 y_labels: {y_labels[:5].flatten()}")
print(f"Last 5 y_labels: {y_labels[-5:].flatten()}")


X_train, X_val, y_train, y_val = train_test_split(
    padded_sequences, y_labels, test_size=0.2, random_state=42
)

print(f"\nShape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of X_val: {X_val.shape}")
print(f"Shape of y_val: {y_val.shape}")


model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("\nModel compiled successfully.")


epochs = 10
batch_size = 32

history = model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=(X_val, y_val)
)

print("\nModel training complete.")
```

```python
import numpy as np


loss, accuracy = model.evaluate(X_val, y_val, verbose=0)

print(f"\nModel Evaluation on Validation Data:")
print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()


plt.tight_layout()


plt.show()
```
